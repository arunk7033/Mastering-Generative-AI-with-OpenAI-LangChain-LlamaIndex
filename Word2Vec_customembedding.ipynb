### Custom training for Word2Vec(static embedding)

```python
import pandas as pd
import numpy as np
import os
import gensim
import nltk
```

```python
nltk.download("all")
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story=[]
```

```python
for file in os.listdir("/content/data"):
  print(file)
```

```python
f=open("/content/data/001ssb.txt",encoding='unicode_escape')
```

```python
corpus=f.read()
```

```python
corpus
```

```python
raw_sentn=sent_tokenize(corpus)
```

```python
for i in raw_sentn:
  print(i)
```

```python
for i in raw_sentn:
  print(simple_preprocess(i))
```

```python
for filename in os.listdir("/content/data"):
  file_path=os.path.join("/content/data",filename)
  with open(file_path,encoding="unicode_escape") as f:
    corpus=f.read()
  raw_sent=sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
story
```

```python
len(story)
```

```python
story[0]
```

```python
story[7]
```

```python
model=gensim.models.Word2Vec(window=10,min_count=5,vector_size=150)
```

```python
model.build_vocab(story)
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.corpus_total_words
```

```python
model.min_count
```

```python
model.train(story,total_examples=model.corpus_count,epochs=5)
```
